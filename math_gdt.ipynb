```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import time


```

```python
L=100
R=1000000
# шаг измерения R, Будет прибавлять на "шаг" к текущему R в цилке
step = 100000

c=[]
```

```python
#функция замера времени подсчета НОД
def nod_time(lst):
    #фиксируем время начала
    start_time = int(round(time.time() * 1000))
    math.gcd(*lst)
    #Фиксируем время конца операции
    end_time = int(round(time.time() * 1000))

    howmtime = end_time - start_time
    print("Затрачено милисекунд на поиск НОД: ", howmtime)
    
    return howmtime
```

```python
results = pd.DataFrame(columns=["R", "gdc"])
results
```

```python
i=0
#посчитаем некоторое количество значений
for i in range(1,200):
    #записываем значения от L до R в список
    c = list(range(L, R))
    #вычисляем скорость вычисления по ранее написанной функции
    gdc_t = nod_time(c)
    
    #записываем значения R и времени на подсчет НОД в Pandas
    new_row = {'R':R,'gdc':gdc_t}
    results = results.append(new_row, ignore_index=True)
    
    #добавляем шаг к R
    R = R + step
    #смотрим на какой итерации i мы сейчас находимся
    print(i)
```

```python
#проверяем данные в таблице
results
```

```python
#нарисуем график, x - значения R, y - значения времени вычисления
plt.figure(figsize=(100, 70))
plt.plot(results['gdc'],results['R'])
plt.show()
```
