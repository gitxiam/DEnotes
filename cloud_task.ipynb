#### Задача 1
Для каждой транзакции посчитать количество дней, которые прошли с момента первой транзакции по этому клиенту. 
На выходе: transaction_id, customer_id, amount_rur, transaction_dttm, days_count.

````python
%%sql

SELECT 
transaction_id, 
customer_id, 
amount_rur, 
transaction_dttm,
DATE_PART('day', transaction_dttm - firts_trn) as days_count

FROM (
	SELECT 
	transaction_id, customer_id, amount_rur, transaction_dttm
    
--если дата первая, значит она должна быть минимальной, самой первой, самой молодой по timestamp
	,MIN(transaction_dttm) OVER(PARTITION BY customer_id ORDER BY transaction_dttm) AS firts_trn 
	FROM transaction
    ) 
    transaction;
````

#### Задача 2
Для каждой транзакции вывести кол-во дней, которые прошли с предыдущей транзакции по клиенту, дату и сумму предыдущей транзакции. 
На выходе: transaction_id, customer_id, amount_rur, transaction_dttm, days_count, prev_transaction_dttm, prev_amount_rur.

````python
%%sql

SELECT 
transaction_id, 
customer_id, 
amount_rur, 
transaction_dttm, 
DATE_PART('day', transaction_dttm - prev_transaction_dttm) as days_count,
prev_transaction_dttm,
prev_amount_rur

FROM (
	SELECT 
	transaction_id, customer_id, amount_rur, transaction_dttm
	,LAG(transaction_dttm) OVER(PARTITION BY customer_id ORDER BY transaction_dttm) AS prev_transaction_dttm
	,LAG(amount_rur) OVER(PARTITION BY customer_id) AS prev_amount_rur
	FROM transaction
    ) 
	transaction;
````

#### Задача 3
Для кaждой трaнзaкции посчитать количество днeй, которые прошли с момента первой транзакции по этому клиенту, и проставить флаг, относится ли транзакция к клиенту, начавшему обслуживаться в банке в текущем году. 
На выходе: transaction_id, customer_id, amount_rur, transaction_dttm, days_count, new_cust_flg.




Покажу сначала подзапрос. В подзапросе я джойню таблицу customer и извлекаю fist_tnrs (первую транзакцию) чтобы на основании этих данных проводить условие простановки флага.

````python
%%sql

SELECT 
transaction.transaction_id, 
transaction.customer_id, 
transaction.amount_rur, 
transaction.transaction_dttm
,MIN(transaction.transaction_dttm) OVER(PARTITION BY transaction.customer_id) AS first_trns 
,customer.start_dttm

FROM transaction 
INNER JOIN customer
ON customer.customer_id=transaction.customer_id;
````

Теперь основной селект:


извлекает количество дней с первой транзакции

```SQL
DATE_PART('day', transaction_dttm - first_trns) as days_count 
```

Простановка флага по принципу: если текущая дата - первая дата = 0, значит TRUE, значит клиент появился в текущем году
Во всех иных случаях ставить false.

```SQL
,case when (DATE_PART('year', CURRENT_DATE) - DATE_PART('year', start_dttm) = 0) THEN 'TRUE'
	else 'false'
	end as new_cust_flg

```

Конечно для предобработки данных имеет смысл указать условие при, котором текущая дата вдруг меньше даты регистрации клиента, но я думаю программисты не дураки, отфиксили этот момент давным давно.

````python
%%sql

SELECT 
transaction_id, customer_id, amount_rur, transaction_dttm 
,DATE_PART('day', transaction_dttm - first_trns) as days_count

,case when (DATE_PART('year', CURRENT_DATE) - DATE_PART('year', start_dttm) = 0) THEN 'TRUE'
	else 'false'
	end as new_cust_flg

FROM 
	(
	SELECT 
	transaction.transaction_id, 
	transaction.customer_id, 
	transaction.amount_rur, 
	transaction.transaction_dttm
	,MIN(transaction.transaction_dttm) OVER(PARTITION BY transaction.customer_id) AS first_trns 
	,customer.start_dttm
	FROM transaction 
	INNER JOIN customer
	ON customer.customer_id=transaction.customer_id
	) transaction;
````
